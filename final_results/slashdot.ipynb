```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
combined_data = pd.read_csv("slashdot.csv")
print(combined_data)
```

```python
# set width of bar
barWidth = 0.1
 
# set height of bar
bars0 = list(combined_data['Base'])
bars1 = list(combined_data['Association'])
bars2 = list(combined_data['Prefetcher'])
bars3 = list(combined_data['Replacement Policy'])
bars4 = list(combined_data['Write Buffer'])
 
# Set position of bar on X axis
r0 = np.arange(len(bars0))
r1 = [x + barWidth for x in r0]
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
 
# Make the plot
plt.bar(r0, bars0, color='black', width=barWidth, edgecolor='white', label='Base')
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='Association')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='Prefetcher')
plt.bar(r3, bars3, color='#0900ff', width=barWidth, edgecolor='white', label='Replacement Policy')
plt.bar(r4, bars4, color='#fff100', width=barWidth, edgecolor='white', label='Write Buffers')
 
# Add xticks on the middle of the group bars
plt.title("Splashdot Dataset\n77360 Nodes \n828161 edges")
plt.ylabel("LLC Miss Rate", fontweight='bold')
plt.xticks([r + barWidth*2 for r in range(len(bars1))], ['BFS', 'CC', 'CC_SV'])
 
# Create legend & Show graphic
plt.legend(loc=(1.04,0))
plt.show()
```
